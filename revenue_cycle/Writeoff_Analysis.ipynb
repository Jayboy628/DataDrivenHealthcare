## SPECIALTY

```python
rpt = writeoff[writeoff['Adjustment Name'].str.contains('Timely Filing')]
rpt = rpt.groupby("Specialty").apply(
    lambda d: pd.Series(
        {
            
            "Distinct Provider": d["Provider Name"].nunique(),
            "Sum of Timely Filing": d["Adjustments"].sum(),
            "TF per Provider":d["Adjustments"].sum() / d["Provider Name"].nunique()
           
          
        }
    )).sort_values(by="TF per Provider", ascending=False)

rpt.loc['Total'] = pd.Series(rpt[["Distinct Provider", "Sum of Timely Filing","TF per Provider"]]
                   .sum(), index = ["Distinct Provider", "Sum of Timely Filing","TF per Provider"])
rpt.style.format({"Distinct Provider": "{:20,.0f}", 
                    "Sum of Timely Filing": "${:20,.2f}", 
                    "TF per Provider": "${:20,.2f}"}).format_index(str.upper, axis=1).background_gradient(cmap='Reds')
```

## DEPARTMENT

```python
rpt1 = writeoff[writeoff['Adjustment Name'].str.contains('Timely Filing')]
rpt1 = rpt1.groupby("Department Name").apply(
    lambda d: pd.Series(
        {
            "Sum of Timely Filing": d["Adjustments"].sum(),
            "Distinct Provider": d["Provider Name"].nunique(),
            "TF Per Provider":d["Adjustments"].sum() / d["Provider Name"].nunique(),
            "YEARLY COST OF TIMELY FILING": d["Adjustments"].sum() * 4
        }
    )).sort_values(by="Sum of Timely Filing", ascending=False)

rpt1.loc['Total'] = pd.Series(rpt1[["Sum of Timely Filing","Distinct Provider", "TF Per Provider", "YEARLY COST OF TIMELY FILING"]]
                 .sum(), index = ["Sum of Timely Filing", "Distinct Provider","TF Per Provider", "YEARLY COST OF TIMELY FILING"])

rpt1.style.format({ "Sum of Timely Filing": "${:20,.2f}", 
                    "Distinct Provider": "{:20,.0f}",
                    "TF Per Provider": "${:20,.2f}",
                    "YEARLY COST OF TIMELY FILING":"${:20,.2f}"}).format_index(str.upper, axis=1).background_gradient(cmap='Reds')
```

## POTENTIAL SAVINGS

```python
rpt1 = writeoff[writeoff['Adjustment Name'].str.contains('Timely Filing')]
rpt1 = rpt1.groupby("Department Name").apply(
    lambda d: pd.Series(
        {
            "Sum of Timely Filing": d["Adjustments"].sum(),
            "Distinct Provider": d["Provider Name"].nunique(),
            "TF Per Provider":d["Adjustments"].sum() / d["Provider Name"].nunique(),
            "YEARLY COST OF TIMELY FILING": d["Adjustments"].sum() * 4
        }
    )).sort_values(by="Sum of Timely Filing", ascending=False).head(5)

rpt1.loc['Total'] = pd.Series(rpt1[["Sum of Timely Filing","Distinct Provider", "TF Per Provider", "YEARLY COST OF TIMELY FILING"]]
                 .sum(), index = ["Sum of Timely Filing", "Distinct Provider","TF Per Provider", "YEARLY COST OF TIMELY FILING"])

rpt1.style.format({ "Sum of Timely Filing": "${:20,.2f}", 
                    "Distinct Provider": "{:20,.0f}",
                    "TF Per Provider": "${:20,.2f}",
                    "YEARLY COST OF TIMELY FILING":"${:20,.2f}"}).format_index(str.upper, axis=1).background_gradient(cmap='Reds')
```
